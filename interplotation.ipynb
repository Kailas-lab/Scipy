```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# Known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])

# Create an interpolation function
f = interpolate.interp1d(x, y)

# Interpolating new points
x_new = np.linspace(0, 5, 50)  # New x values
y_new = f(x_new)  # Interpolated y values

# Plot original data and interpolated values
plt.plot(x, y, 'o', label='Original Data')  # Original points
plt.plot(x_new, y_new, '-', label='Interpolated Curve')  # Interpolated curve
plt.legend()
plt.show()
```
